```python
import torch
import cv2

import shutil
shutil.rmtree("C:/Users/watee/.cache/torch/hub", ignore_errors=True)

import pathlib
temp = pathlib.PosixPath
pathlib.PosixPath = pathlib.WindowsPath


model = torch.hub.load("ultralytics/yolov5", "custom", path="best.pt", force_reload=True)


video = cv2.VideoCapture(r"C:\Users\watee\OneDrive\Desktop\video project.MP4")

while True:
    ret, frame = video.read()
    if not ret:
        break

  
    results = model(frame)

   
    frame_with_boxes = results.render()[0]
    

   
    cv2.imshow("Saudi Plant Safety detector - Video Test", frame_with_boxes)

    if cv2.waitKey(1) == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```
